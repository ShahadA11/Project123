2) Create a script to develop a Python function for descriptive statistics. The input for the function should be the sample and the field to perform the descriptive statistics

```python
import pandas as pd

def Desc_stat(ds, var):
    mean = ds[var].mean()
    median = ds[var].median()
    mode = ds[var].mode()[0]
    minimum = ds[var].min()
    maximum = ds[var].max()
    rangee = ds[var].max() - ds[var].min()
    std_dev = ds[var].std()
    variance = ds[var].var()
    skew = ds[var].skew()
    kurtosis = ds[var].kurt()
    count = ds[var].count()

    Quart = [
        ds[var].quantile(0),
        ds[var].quantile(0.25),
        ds[var].quantile(0.50),
        ds[var].quantile(0.75),
        ds[var].quantile(1),
        ds[var].quantile(0.75) - ds[var].quantile(0.25)
    ]

    summary = {
        "Average": mean,
        "Median": median,
        "Mode": mode,
        "Minimum": minimum,
        "Maximum": maximum,
        "Range": rangee,
        "Std Dev": std_dev,
        "Variance": variance,
        "Skewness": skew,
        "Kurtosis": kurtosis,
        "Minimum (Q0)": Quart[0],
        "25th Percentile (Q1)": Quart[1],
        "50th Percentile (Q2)": Quart[2],
        "75th Percentile (Q3)": Quart[3],
        "Maximum (Q4)": Quart[4],
        "Inter Quartile Range (IQR)": Quart[5],
        "Count": count
    }

    return summary
```

```python
# Descriptive statistics function
def custom_descriptive_stats(series):
    stats = {
        'Count': series.count(),
        'Mean': series.mean(),
        'Median': series.median(),
        'Min': series.min(),
        'Max': series.max(),
        'Std Dev': series.std(),
        'Variance': series.var(),
        'Skewness': series.skew(),
        'Kurtosis': series.kurt(),
        '25% Quantile': series.quantile(0.25),
        '50% Quantile': series.quantile(0.5),
        '75% Quantile': series.quantile(0.75)
    }
    return stats

# Systematic sampling function
def systematic_sampling(df, sample_size):
    step = len(df) // sample_size
    start = np.random.randint(0, step)
    indexes = np.arange(start, len(df), step)
    sample = df.iloc[indexes[:sample_size]]
    return sample

# Create systematic sample
sample_size = 150
systematic_sample = systematic_sampling(data, sample_size)

# Get descriptive stats for 'Caffeine_mg'
caffeine_stats = custom_descriptive_stats(systematic_sample['Caffeine_mg'])

# Show results
print("\nDescriptive Statistics for 'Caffeine_mg' (Systematic Sample of 150 rows):\n")
for key, value in caffeine_stats.items():
    print(f"{key:25}: {value:.2f}")
```

```python
# Import all needed libraries
import numpy as np
from scipy.stats import pearsonr, spearmanr, chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Choose variables
# Numerical independent variable: Sleep_Hours
# Numerical dependent variable: Caffeine_mg
# Categorical independent variable: Smoking

# Step 2: Pearson and Spearman correlation between Sleep_Hours and Caffeine_mg
pearson_corr, pearson_pval = pearsonr(data['Sleep_Hours'], data['Caffeine_mg'])
spearman_corr, spearman_pval = spearmanr(data['Sleep_Hours'], data['Caffeine_mg'])

print("Numerical Variables: Sleep_Hours vs Caffeine_mg")
print(f"Pearson Correlation: {pearson_corr:.3f}, P-value: {pearson_pval:.3f}")
print(f"Spearman Correlation: {spearman_corr:.3f}, P-value: {spearman_pval:.3f}")

# Step 3: Chi-Square test between Smoking (categorical) and Caffeine_mg
# First, group Caffeine_mg into categories (Low, Medium, High)
data['Caffeine_cat'] = pd.qcut(data['Caffeine_mg'], q=3, labels=["Low", "Medium", "High"])

# Create contingency table
contingency_table = pd.crosstab(data['Smoking'], data['Caffeine_cat'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nCategorical Variables: Smoking vs Caffeine Category")
print(f"Chi-Square Statistic: {chi2:.3f}")
print(f"P-value: {p:.3f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:")
print(expected)

# Step 4: Visualize relationship with a heatmap
sns.heatmap(contingency_table, annot=True, cmap='Blues')
plt.title('Heatmap of Smoking Status vs Caffeine Category')
plt.xlabel('Caffeine Category')
plt.ylabel('Smoking Status')
plt.show()
```

```python
# Import the needed library
from scipy.stats import ttest_1samp
import numpy as np

# Step 1: Find the population mean of Caffeine_mg
population_mean = data['Caffeine_mg'].mean()
print(f"Population Mean of Caffeine (mg): {population_mean:.2f}")

# Step 2: Take a random sample of 150 records
sample = data['Caffeine_mg'].sample(n=150, random_state=42)

# Step 3: Perform one-sample t-test
# Compare the sample mean with the population mean
t_statistic, p_value = ttest_1samp(sample, population_mean)

print("\nOne-Sample T-Test Result:")
print(f"T-Statistic: {t_statistic:.3f}")
print(f"P-Value: {p_value:.3f}")

# Step 4: Conclusion
alpha = 0.05  # 5% significance level
if p_value < alpha:
    print("\nResult: Reject the null hypothesis.")
    print("The sample mean is significantly different from the population mean.")
else:
    print("\nResult: Fail to reject the null hypothesis.")
    print("The sample mean is not significantly different from the population mean.")
```

```python
# Import all libraries
import pandas as pd
from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler

# Step 1: Load the dataset
data = pd.read_csv('Global health coffee.csv')

# Step 2: Create a new column for high or low caffeine
data['High_Caffeine'] = (data['Caffeine_mg'] > data['Caffeine_mg'].median()).astype(int)

# Step 3: Remove the original caffeine column
data.drop(['Caffeine_mg'], axis=1, inplace=True)

# Step 4: Change text columns into numbers
data = pd.get_dummies(data, drop_first=True).astype(int)

# Step 5: Split into input (X) and output (y)
X = data.drop('High_Caffeine', axis=1)
y = data['High_Caffeine']

# Step 6: Scale all input values (for better model performance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 7: Use 5 parts for cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Step 8: Create a dictionary to store all model results
cv_results = {}

# Logistic Regression
logistic_model = LogisticRegression(max_iter=1000, random_state=42)
logistic_y_pred_cv = cross_val_predict(logistic_model, X_scaled, y, cv=cv)
logistic_conf_matrix_cv = confusion_matrix(y, logistic_y_pred_cv)
logistic_accuracy_cv = accuracy_score(y, logistic_y_pred_cv)
logistic_errors_cv = logistic_conf_matrix_cv[0,1] + logistic_conf_matrix_cv[1,0]
cv_results['Logistic Regression'] = {
    'Confusion Matrix': logistic_conf_matrix_cv,
    'Errors': logistic_errors_cv,
    'Accuracy': logistic_accuracy_cv,
    'Predictions': logistic_y_pred_cv
}

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_y_pred_cv = cross_val_predict(knn_model, X_scaled, y, cv=cv)
knn_conf_matrix_cv = confusion_matrix(y, knn_y_pred_cv)
knn_accuracy_cv = accuracy_score(y, knn_y_pred_cv)
knn_errors_cv = knn_conf_matrix_cv[0,1] + knn_conf_matrix_cv[1,0]
cv_results['K-Nearest Neighbors'] = {
    'Confusion Matrix': knn_conf_matrix_cv,
    'Errors': knn_errors_cv,
    'Accuracy': knn_accuracy_cv,
    'Predictions': knn_y_pred_cv
}

# Naive Bayes
nb_model = GaussianNB()
nb_y_pred_cv = cross_val_predict(nb_model, X_scaled, y, cv=cv)
nb_conf_matrix_cv = confusion_matrix(y, nb_y_pred_cv)
nb_accuracy_cv = accuracy_score(y, nb_y_pred_cv)
nb_errors_cv = nb_conf_matrix_cv[0,1] + nb_conf_matrix_cv[1,0]
cv_results['Naive Bayes'] = {
    'Confusion Matrix': nb_conf_matrix_cv,
    'Errors': nb_errors_cv,
    'Accuracy': nb_accuracy_cv,
    'Predictions': nb_y_pred_cv
}

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_y_pred_cv = cross_val_predict(dt_model, X_scaled, y, cv=cv)
dt_conf_matrix_cv = confusion_matrix(y, dt_y_pred_cv)
dt_accuracy_cv = accuracy_score(y, dt_y_pred_cv)
dt_errors_cv = dt_conf_matrix_cv[0,1] + dt_conf_matrix_cv[1,0]
cv_results['Decision Tree'] = {
    'Confusion Matrix': dt_conf_matrix_cv,
    'Errors': dt_errors_cv,
    'Accuracy': dt_accuracy_cv,
    'Predictions': dt_y_pred_cv
}

# Step 9: Create a table to show all results
cv_formatted_results = []
for model_name, metrics in cv_results.items():
    cv_formatted_results.append({
        'Model': model_name,
        'Confusion Matrix': metrics['Confusion Matrix'],
        'Errors (FP + FN)': metrics['Errors'],
        'Accuracy (%)': round(metrics['Accuracy'] * 100, 2),
        'First 5 Predictions': metrics['Predictions'][:5].tolist()
    })

# Step 10: Show final results
cv_results_df = pd.DataFrame(cv_formatted_results)
print("\nCross Validation Results:\n")
print(cv_results_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

# Step 1: Load dataset
dataset = pd.read_csv('Global health coffee.csv')

# Step 2: Select fields for clustering
X = dataset[['BMI', 'Caffeine_mg']].values
print(X)

# Step 3: Find the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Step 4: Apply K-Means with the best number of clusters (3)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)

# Step 5: Visualize the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='yellow', label='Centroids')

plt.title('Clusters of Individuals (K-Means)')
plt.xlabel('BMI')
plt.ylabel('Caffeine (mg)')
plt.legend()
plt.show()
```

```python
# (Hierarchical / Horizontal Clustering)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

# Step 1: Load dataset
dataset = pd.read_csv('Global health coffee.csv')

# Step 2: Select fields for clustering
X = dataset[['BMI', 'Caffeine_mg']].values
print(X)

# Step 3: Create Dendrogram
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Individuals')
plt.ylabel('Euclidean Distances')
plt.show()

# Step 4: Fit Hierarchical Clustering
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
y_hc = hc.fit_predict(X)

# Step 5: Visualize the clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s=100, c='green', label='Cluster 3')

plt.title('Clusters of Individuals (Hierarchical Clustering)')
plt.xlabel('BMI')
plt.ylabel('Caffeine (mg)')
plt.legend()
plt.show()
```
