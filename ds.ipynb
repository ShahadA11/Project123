```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Global health coffee.csv')

#Take a random sample of 150 rows
sample = data.sample(n=150, random_state=42)

#Define a function for descriptive statistics
def custom_descriptive_stats(series):
    stats = {
        'Count': series.count(),
        'Mean': series.mean(),
        'Median': series.median(),
        'Min': series.min(),
        'Max': series.max(),
        'Standard Deviation': series.std(),
        'Variance': series.var(),
        'Skewness': series.skew(),
        'Kurtosis': series.kurt(),
        '25% Quantile': series.quantile(0.25),
        '50% Quantile (Median)': series.quantile(0.5),
        '75% Quantile': series.quantile(0.75)
    }
    return stats

# Apply the function to the dependent variable 'Caffeine_mg'
caffeine_stats = custom_descriptive_stats(sample['Caffeine_mg'])

#Display the result
print("\nDescriptive Statistics for 'Caffeine_mg' (Sample of 150 rows):\n")
for key, value in caffeine_stats.items():
    print(f"{key:30}: {value:.2f}")
```
